<span style = "font-size: 20px;"> The columns Amount, Status and Unnamed1 have null values. </span>

```python
df[df.Amount.isnull()]
```

```python
#Drop unnecessary columns which are not required.
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

Unnecessary columns have been dropped

```python
#Drop the entire row containing NA or NULL values. inplace = True ensures that the change is complete.
df.dropna(inplace=True)
```

```python
df.shape
```

Number of rows changed from 11251 to 11239, confirming that 12 rows have been dropped.

```python
df['Amount'] = df['Amount'].astype('int')

#Change the datatype from float to int.
```

```python
df.rename(columns = {'Cust_name': 'Customer_name', 'User_ID': 'UserID', 'Product_ID': 'ProductID'}, inplace = True)

#Renaming columns for uniformity and readability.
```

```python
df['Gender'] = df['Gender'].map({'F': 'Female', 'M': 'Male'})
```

```python
df.drop_duplicates()
```

There are no duplicate rows in the dataset, as the no. of rows remains the same. Dataset has been cleaned and is ready for EDA and further analysis.

```python
df.to_csv('cleaned_diwali_sales.csv', index=False)
```
